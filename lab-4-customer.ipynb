```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
numerical_columns = ['Customer Lifetime Value', 'Income', 'Monthly Premium Auto',
                      'Months Since Last Claim', 'Months Since Policy Inception',
                      'Number of Open Complaints', 'Number of Policies', 'Total Claim Amount']

sns.set(style="whitegrid")

fig, axes = plt.subplots(nrows=2, ncols=4, figsize=(18, 10))
fig.suptitle('Distribution Plots for Numerical Columns', y=1.02)

for i, column in enumerate(numerical_columns):
    row = i // 4
    col = i % 4
    sns.histplot(df[column], kde=True, ax=axes[row, col], color='skyblue')
    axes[row, col].set_title(column)

plt.tight_layout()
plt.show()
```

```python
numerical_data = df[numerical_columns]

correlation_matrix = numerical_data.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix for Numerical Variables')
plt.show()
```

```python
numerical_data = df[numerical_columns]


correlation_matrix = numerical_data.corr()


highly_correlated_pairs = [(i, j) for i in range(correlation_matrix.shape[0])
                           for j in range(i + 1, correlation_matrix.shape[1])
                           if abs(correlation_matrix.iloc[i, j]) > 0.9]


if len(highly_correlated_pairs) > 0:
    print("Highly correlated pairs:")
    for pair in highly_correlated_pairs:
        feature1, feature2 = numerical_columns[pair[0]], numerical_columns[pair[1]]
        correlation_value = correlation_matrix.iloc[pair[0], pair[1]]
        print(f"{feature1} and {feature2}: {correlation_value}")

    
    for pair in highly_correlated_pairs:
        feature1, feature2 = numerical_columns[pair[0]], numerical_columns[pair[1]]
        
        if correlation_matrix.loc[feature1, :].abs().sum() > correlation_matrix.loc[feature2, :].abs().sum():
            numerical_data = numerical_data.drop([feature1], axis=1)
        else:
            numerical_data = numerical_data.drop([feature2], axis=1)

    
    correlation_matrix_after_drop = numerical_data.corr()

    
    plt.figure(figsize=(12, 10))
    sns.heatmap(correlation_matrix_after_drop, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
    plt.title('Updated Correlation Matrix for Numerical Variables (After Dropping)')
    plt.show()

else:
    print("No pairs of features with high correlation (> 0.9) found.")
```
